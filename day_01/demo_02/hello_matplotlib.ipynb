```python
# 0.
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1, 4) for i in x]

# 1.
plt.figure(figsize=(20, 8), dpi=100)

# 2.
plt.plot(x, y_shanghai, label="")
plt.plot(x, y_beijing, color="r", linestyle="--", label="")

# xy
x_ticks_label = ["11{}".format(i) for i in x]
y_ticks = range(40)

# xy
plt.xticks(x[::5], x_ticks_label[::5])
plt.yticks(y_ticks[::5])

# 
plt.grid(True, linestyle="--", alpha=1)

# 

plt.xlabel("", fontsize=10)
plt.ylabel("", fontsize=10)
plt.title("11-12", fontsize=20)

# 
plt.savefig("./test.png")

# 
plt.legend(loc="best")

# 3.
plt.show()
```

```python
# 0.
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1, 4) for i in x]

# 1.
# plt.figure(figsize=(20, 8), dpi=100)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), dpi=100)

# 2.
# plt.plot(x, y_shanghai, label="")
# plt.plot(x, y_beijing, color="r", linestyle="--", label="")
axes[0].plot(x, y_shanghai, label="")
axes[1].plot(x, y_beijing, color="r", linestyle="--", label="")

# xy
x_ticks_label = ["11{}".format(i) for i in x]
y_ticks = range(40)

# # xy
# plt.xticks(x[::5], x_ticks_label[::5])
# plt.yticks(y_ticks[::5])

axes[0].set_xticks(x[::5], x_ticks_label[::5])
axes[0].set_yticks(y_ticks[::5])
# axes[0].set_xticklabels(x_ticks_label[::5])


# # 
# plt.grid(True, linestyle="--", alpha=1)
axes[0].grid(True, linestyle="--", alpha=0.5)
axes[1].grid(True, linestyle="--", alpha=0.5)


# # 
# plt.xlabel("", fontsize=10)
# plt.ylabel("", fontsize=10)
# plt.title("11-12", fontsize=20)
axes[0].set_xlabel("", fontsize=10)
axes[0].set_ylabel("", fontsize=10)
axes[0].set_title("11-12", fontsize=20)

axes[1].set_xlabel("", fontsize=10)
axes[1].set_ylabel("", fontsize=10)
axes[1].set_title("11-12", fontsize=20)

# # 
plt.savefig("./test.png")

# # 
# plt.legend(loc="best")
axes[0].legend(loc="best")
axes[1].legend(loc="best")

# 3.
plt.show()
```
