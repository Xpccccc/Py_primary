## 生成固定范围的数组

```python
np.linspace(10,100,11)  # 生成11个元素的等差数列
```

```python
np.arange(10,1000,9)  # 生成以2为公差的从10到100（不包含100）等差数列
```

```python
np.logspace(0, 6, 4, dtype=int) # 生成从10^0 ~ 10^6的 4个等比数列 
```

## 生成随机数组

### 正态分布

```python
x1 = np.random.normal(1.75, 1, 100000000) # 生成均值为1.75，标准差为1的 100000000个数据
```

```python
x1
```

```python
# 1.创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 2.绘制图像
plt.hist(x1, 1000) # 1000表示分成1000组

# 3.显示图像
plt.show()
```

### 均匀分布

```python
x2 = np.random.uniform(-1,1,100000000) # 生成值为 -1～1 的 100000000个数据
```

```python
x2
```

```python
# 1.创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 2.绘制图像
plt.hist(x2, 1000) # 1000表示分成1000组

# 3.显示图像
plt.show()
```
